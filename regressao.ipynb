```python
import pandas as pd
import numpy as np
import random

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

%matplotlib inline
```

```python
df = pd.read_csv('INPI.csv', sep=';', decimal=',')
```

```python
df.head()
```

```python
for i in range(1,4):
    t_menos_1 = 'temp_t_{}'.format(i)
    df[t_menos_1] = df.TempAr.shift(i)
    t_mais_1 = 'temp_tm{}'.format(i)
    df[t_mais_1] = df.TempAr.shift(-i)
```

```python
df
```

```python
for i in range(3,22,3):
    desc_h = 'h{}'.format(i*100)
    df[desc_h] = df.HORA.apply(lambda x: 1 if x == i*100 else 0)
```

```python
for i in range(1,12):
    desc_m = 'm{}'.format(i)
    df[desc_m] = df.HORA.apply(lambda x: 1 if x == i else 0)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
model = linear_model.LinearRegression()
```

```python
y = df.TempAr.values.reshape(-1,1)
```

```python
df.columns[5:]
```

```python
x = df[df.columns[5:]].values
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x)
```

```python
r2_score(y, y_pred)
```
